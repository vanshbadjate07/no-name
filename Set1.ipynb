**# Q1 Identify and fix the architectural error. Write corrected Keras model code with the proper output activation for 10-class classification and the correct loss function in model.compile().**

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Corrected model for 10-class classification
model = Sequential([
    Dense(128, activation='relu', input_shape=(784,)),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')  # Fixed: softmax for 10-class
])

model.compile(optimizer='adam',
              loss='categorical_crossentropy',  # Fixed: correct loss function
              metrics=['accuracy'])
```

**# Q2 Write Python code using TensorFlow to compute the Categorical Cross-Entropy loss manually for a sample prediction vector [0.1, 0.7, 0.2] against true label index 1, and print the result.**

```python
import tensorflow as tf

# Prediction probabilities and true label
y_pred = tf.constant([0.1, 0.7, 0.2])
y_true_idx = 1

# Manual computation of categorical cross-entropy loss
loss = -tf.math.log(y_pred[y_true_idx])
print("Categorical Cross-Entropy Loss:", loss.numpy())
```
