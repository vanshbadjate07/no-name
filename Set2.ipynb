**# Q1 Write Keras code to build a Dense network with Dropout(0.4) after each hidden layer. Compile and train it on a small dataset (use make_classification from sklearn to generate data).**

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Dropout

# Generate data
X, y = make_classification(n_samples=1000, n_features=20, random_state=42)

# Split data
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2)

# Build model
model = Sequential([
    Dense(128, activation='relu', input_shape=(20,)),
    Dropout(0.4),
    Dense(64, activation='relu'),
    Dropout(0.4),
    Dense(1, activation='sigmoid')
])

# Compile
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train
model.fit(X_train, y_train, epochs=20, validation_data=(X_val, y_val))
```

**# Q2 Modify the same network to add L2 kernel_regularizer (lambda=0.001) to each Dense layer. Train both models and print both validation accuracies to compare the regularization effect.**

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras import regularizers

# Generate data
X, y = make_classification(n_samples=1000, n_features=20, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2)

# Dropout Model
model_drop = Sequential([
    Dense(128, activation='relu', input_shape=(20,)),
    Dropout(0.4),
    Dense(64, activation='relu'),
    Dropout(0.4),
    Dense(1, activation='sigmoid')
])

model_drop.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
history_drop = model_drop.fit(X_train, y_train, epochs=20, validation_data=(X_val, y_val), verbose=0)

# L2 Model
model_l2 = Sequential([
    Dense(128, activation='relu', input_shape=(20,), kernel_regularizer=regularizers.l2(0.001)),
    Dense(64, activation='relu', kernel_regularizer=regularizers.l2(0.001)),
    Dense(1, activation='sigmoid')
])

model_l2.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
history_l2 = model_l2.fit(X_train, y_train, epochs=20, validation_data=(X_val, y_val), verbose=0)

# Compare results
drop_acc = history_drop.history['val_accuracy'][-1]
l2_acc = history_l2.history['val_accuracy'][-1]

print("Dropout Val Accuracy:", drop_acc)
print("L2 Val Accuracy:", l2_acc)
```
